```python
import numpy as np
import pandas as pd
import random
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import load_breast_cancer
c1 = load_breast_cancer()
df = pd.DataFrame(c1['data'],columns=c1['feature_names'])
label = c1['target']
```

```python
df
```

```python
df.shape, label.shape
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(df, label, test_size = 0.3, random_state = 12)
```

```python
x_train.shape
```

# First train using Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
LM = LogisticRegression()
LM.fit(x_train, y_train)
LM_pred = LM.predict(x_test)
print(accuracy_score(y_test,LM_pred))
```

# Apply GA to solve the same Problem

```python
# Initialize Population
pop = []
for i in range(200):      # pop_size = 10
    chromosome = np.ones(30, dtype = np.intc)  # num_features = 5
    chromosome[:int(0.3*30)] = False
    np.random.shuffle(chromosome)
    pop.append(chromosome)
```

```python
chromosome
```

```python
pop
```

# We have to select two sets of chromosomes as parents...
## We need Fitness Function and its output as scores

```python
scores = []
for chromosome in pop:
    LM.fit(x_train.iloc[:,chromosome],y_train)
    LM_predt = LM.predict(x_test.iloc[:,chromosome])
    scores.append(accuracy_score(y_test,LM_predt))
scores, pop = np.array(scores), np.array(pop)
```

```python
x_train.iloc[:,chromosome].shape
```

```python
inds = np.argsort(scores)
```

```python
inds
```

```python
list(scores[inds][::-1])
```

```python
list(pop[inds,:][::-1])
```

```python
pop_after_fit = list(pop[inds,:][::-1])
```

# Select Next Gen Pop

```python
pop_nextgen = []
for i in range(100):
    pop_nextgen.append(pop_after_fit[i])
```

```python
len(pop_nextgen),pop_nextgen
```

# Crossover Operation

```python
pop_nextgeneration = pop_nextgen
for i in range(len(pop_nextgen)):
    child = pop_nextgen[i]
    child[3:7] = pop_nextgen[(i+1)%len(pop_nextgen)][3:7]
    pop_nextgeneration.append(child)
```

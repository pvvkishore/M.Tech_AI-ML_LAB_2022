```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Latest Covid-19 India Status.csv')
```

```python
df
```

```python
plt.figure(figsize=(7,7))
plt.scatter(df["Active Ratio (%)"],df["Death Ratio (%)"])
plt.xlabel('Active Ratio (%)')
plt.ylabel('Death Ratio (%)')
plt.title('Data Distribution')
```

```python
X = df[['Active Ratio (%)', 'Death Ratio (%)']] 
```

```python
from sklearn.mixture import GaussianMixture
gmm = GaussianMixture(n_components = 4)
```

```python
gmm.fit(X)
```

```python
Labels = gmm.predict(X)
```

```python
df_r = pd.DataFrame(X)
```

```python
df_r['cluster'] = Labels
df_r
```

```python
df_r.columns = ['Active Ratio (%)', 'Death Ratio (%)', 'cluster']
```

```python
color = ['green', 'red', 'blue','cyan']
```

```python
for k in range(0,4):
    X = df_r[df_r["cluster"]==k]
    plt.scatter(X["Active Ratio (%)"], X["Death Ratio (%)"],c = color[k])
```

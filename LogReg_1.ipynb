```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from math import exp
plt.rcParams["figure.figsize"] = (10, 6)
```

```python
data = pd.read_csv('dataM.csv')
```

```python
data
```

```python
X = data.iloc[:,0]
Y = data.iloc[:,1]
```

```python
Y
```

```python
plt.scatter(X,Y)
```

```python
def predict(X, b0, b1):
    return np.array([1 / (1 + exp(-b0 -b1*x)) for x in X])
```

```python
b0 = 0
b1 = 0
L = 0.001
epochs = 100
for epoch in range(epochs):
    y_pred = predict(X, b0, b1)
    D_b0 = (-2/len(X)) * sum((Y - y_pred) * y_pred * (1 - y_pred))  # Derivative of loss wrt b0
    D_b1 = (-2/len(X)) * sum(X * (Y - y_pred) * y_pred * (1 - y_pred))  # Derivative of loss wrt b1
    # Update b0 and b1
    b0 = b0 - L * D_b0
    b1 = b1 - L * D_b1
    print(epoch)
    print(y_pred)
```

```python
X = df1.iloc[:,:4]
X
```

```python
y = df1.iloc[:,4]
y
```

# Encode the 'STATE' column where we have one hot encoding values

```python
from sklearn.preprocessing import OneHotEncoder
Ohe = OneHotEncoder(sparse = False)
x = Ohe.fit_transform(df1[['State']])
x
```

```python
Ohe.categories_
```

```python
from sklearn.compose import make_column_transformer
col_transformer = make_column_transformer(
    (OneHotEncoder(handle_unknown = 'ignore'),['State']),
    remainder = 'passthrough')
x = col_transformer.fit_transform(X)
x.shape
x
```

### Split the Dataset into Train Set and Test Set

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
#shapes of splitted data
print("X_train:",x_train.shape)
print("X_test:",x_test.shape)
print("Y_train:",y_train.shape)
print("Y_test:",y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
linreg_multi=LinearRegression()
linreg_multi.fit(x_train,y_train)
```

```python
y_pred=linreg_multi.predict(x_test)
y_pred
```

# Evaluate the Model

```python
Accuracy=r2_score(y_test,y_pred)*100
print(" Accuracy of the model is %.2f" %Accuracy)
```

# Plot the Results

```python
plt.scatter(y_test,y_pred);
plt.xlabel('Actual');
plt.ylabel('Predicted');
```

```python
sns.regplot(x=y_test,y=y_pred,ci=None,color ='red');
```

# Predicted Versus Target Values

```python
pred_df=pd.DataFrame({'Actual Value':y_test,'Predicted Value':y_pred,'Difference':y_test-y_pred})
```

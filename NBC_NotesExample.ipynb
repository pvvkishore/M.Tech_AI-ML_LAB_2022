# NB Classifier For Our Class Example

# Define Dataset

```python
outlook = ['rainy','rainy','overcast','sunny','sunny','sunny','overcast','rainy','rainy','sunny','rainy','overcast','overcast','sunny']
```

```python
temp = ['hot','hot','hot','mild','cool','cool','cool','mild','cool','mild','mild','mild','hot','mild']
```

```python
type(temp)
```

```python
len(outlook)
```

```python
humidity = ['high','high','high','high','normal','normal','normal','high','normal','normal','normal','high','normal','high']
```

```python
len(humidity)
```

```python
windy = ['false','true','false','false','false','true','true','false','false','false','true','true','false','true']
```

```python
len(windy)
```

```python
play_golf = ['no','no','yes','yes','yes','no','yes','no','yes','yes','yes','yes','yes','no']
```

```python
len(play_golf)
```

```python
from sklearn import preprocessing
from sklearn.naive_bayes import GaussianNB
```

```python
le = preprocessing.LabelEncoder()
outlook_encoded=le.fit_transform(outlook)
temp_encoded=le.fit_transform(temp)
humidity_encoded=le.fit_transform(humidity)
windy_encoded=le.fit_transform(windy)
```

```python
print(temp_encoded),print(temp)
```

```python
label=le.fit_transform(play_golf)
```

```python
label,play_golf
```

```python
features=list(zip(outlook_encoded,temp_encoded,humidity_encoded,windy_encoded))
```

```python
features
```

```python
label
```

```python
model = GaussianNB()
```

```python
model.fit(features,label)
```

```python
predicted= model.predict([[1,2,0,1]]) # 0:Overcast, 2:cool, 1.Normal,1,true
print("Predicted Value:", predicted)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
a_s = pd.read_csv('apple_strawberry.csv',header=None)
```

```python
a_s.values[1]
```

```python
n_samples, n_features = a_s.shape
print(n_samples)
print(n_features)
```

```python
for i in range(15):
    img = a_s.values[i].reshape((64,64))
    plt.imshow(img, cmap="Greys")
    plt.show()
```

```python
from sklearn import decomposition
pca = decomposition.PCA(n_components=10, whiten=True)
pca.fit(a_s)
```

```python
print(pca.components_)
```

```python
fig = plt.figure(figsize=(16, 6))
for i in range(10):
    ax = fig.add_subplot(2, 5, i + 1, xticks=[], yticks=[])
    #img = pca.components_[i].reshape((64,64))
    ax.imshow(pca.components_[i].reshape((64,64)), cmap="Greys")
    #plt.show()
```

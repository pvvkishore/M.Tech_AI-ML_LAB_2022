# Y = m1x1+m2x2+m3x3+c+bias, Now what is bias.

```python
L = 0.001
```

# Learning Rate....Change L and see for your self

```python
epochs = 100
```

```python
a = 0
```

```python
b = 0
```

```python
n = float(len(X))
```

```python
n
```

# n = number of elements in input X

# Now Build the model to Train

```python
for i in range(epochs):
    Y_pred = a*X+b
    d_a = (-2/n)*sum(X*(Y-Y_pred))
    d_b = (-2/n)*sum(Y-Y_pred)
    a = a - L*d_a
    b = b - L*d_b
    plt.scatter(X,Y)
    plt.plot([min(X),max(X)],[min(Y_pred),max(Y_pred)],color='red')
    plt.show()
    print(i)
    print(a,b)
```
